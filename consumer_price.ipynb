```python
import pandas as pd
```

```python
df = pd.read_csv('Resource/Processed Data/brics_us_consumer_inflation.csv')
df
```

```python
df=df.dropna()
```

```python
brics_us=df.set_index(['Country Code'])

brics_us
```

```python
brics_us = brics_us.drop(columns=['Indicator Name','Indicator Code'])
brics_us
```

```python
import matplotlib.pyplot as plt
from matplotlib import style

brics_us = brics_us.T
brics_us
```

```python
brics_us = brics_us.drop(index=['Country Name'])
```

```python
brics_us
```

```python
brics_us.plot(figsize=(15,10))
style.use('fivethirtyeight')
plt.title("Consumer Price Index by Country")
plt.xlabel("Year")
plt.ylabel("CPI")
```
