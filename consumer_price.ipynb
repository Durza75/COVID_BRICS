```python
import pandas as pd
```

```python
df = pd.read_csv('Resource/brics_us_consumer_inflation.csv')
df
```

```python
brics_us = df.set_index('Country Name')
brics_us 
```

```python
brics_us = brics_us.groupby(['Country Name']).sum()
brics_us
```

```python
import matplotlib.pyplot as plt
from matplotlib import style

brics_us = brics_us.T
brics_us
```

```python
brics_us.plot(figsize=(15,10))
style.use('fivethirtyeight')
plt.title("Consumer Price Index by Country")
plt.xlabel("Year")
plt.ylabel("CPI")
```
